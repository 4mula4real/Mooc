### Use Monte Carlo algorithms to calculate pi directly

```python
def direct_pi(n_trials, r):
    n_hits = 0
    for i in range(n_trials):
        x = random.uniform(-r, r)
        y = random.uniform(-r, r)
        if x ** 2 + y ** 2 <= r**2:
            n_hits += 1
    return n_hits

n_trials = 5000
r = 1
prob = direct_pi(n_trials, r) / n_trials
pi = (2 * r) ** 2 * prob
pi
```

```python
# mulitirun direct_pi
n_runs = 1000
pi_sum = 0
for run in range(n_runs):
    prob = direct_pi(n_trials, r) / n_trials
    pi = (2 * r) ** 2 * prob
    pi_sum += pi
    print(pi)

print("average pi = " + str(pi_sum/n_runs))
```

### Markov chain Monte Carlo(MCMC)

(进入每一个点的可能性都一样)

```python
def markov_pi(n_trials, delta):
    x = 1.0
    y = 1.0
    n_hits = 0
    for i in range(n_trials):
        del_x = random.uniform(-delta, delta)
        del_y = random.uniform(-delta, delta)
        if abs(x + del_x) < 1.0 and abs(y + del_y) < 1.0: # if in the square
            x += del_x
            y += del_y
        if x**2 + y**2 < 1.0: # if in the circle
            n_hits += 1
    return n_hits


delta = 0.1
prob = direct_pi(n_trials, r) / n_trials
pi = (2 * r) ** 2 * prob
pi
```

```python
# mulitirun markov_pi
n_runs = 1000
pi_sum = 0
for run in range(n_runs):
    prob = markov_pi(n_trials, delta) / n_trials
    pi = (2 * r) ** 2 * prob
    pi_sum += pi
    print(pi)

print("average pi = " + str(pi_sum/n_runs))
```

### Basic pebble game

```python
# =============================================================================
# The var "neighbor" corresponds to the possible move for below 3 x 3 matrix
# 6 7 8
# 3 4 5
# 0 1 2
# (upside down)
# =============================================================================

neighbor =  [[1, 3, 0, 0], [2, 4, 0, 1], [2, 5, 1, 2],
             [4, 6, 3, 0], [5, 7, 3, 1], [5, 8, 4, 2],
             [7, 6, 6, 3], [8, 7, 6, 4], [8, 8, 7, 5]]

t_max = 4 # max time
site = 8 # starting point
t = 0
print(site)
while t < t_max:
    t += 1
    site = neighbor[site][random.randint(0, 3)]
    print(site)
```

```python
# pebble_basic_multirun
t_max = 4 # max time
N_runs = 4**t_max * 100
count_4 = 0
for run in range(N_runs):
    site = 8 # starting point
    t = 0
    while t < t_max:
        t += 1
        site = neighbor[site][random.randint(0, 3)]
    if site == 4:
        count_4 += 1        
    print(site)
    
print("The empirical probability of ending in 4 = " + str(count_4/N_runs))
```

```python
#Inhomogeneous pebble game
histo = [0, 0, 0, 0, 0, 0, 0, 0, 0]
neighbor =  [[1, 3, 0, 0], [2, 4, 0, 1], [2, 5, 1, 2],
             [4, 6, 3, 0], [5, 7, 3, 1], [5, 8, 4, 2],
             [7, 6, 6, 3], [8, 7, 6, 4], [8, 8, 7, 5]]
weight = [3.0, 0.5, 1.0, 0.5, 1.0, 0.5, 2.0, 0.5, 1.0]
pos = 8
n_iter = 1000000


for iter in range(n_iter):
    new_pos = neighbor[pos][random.randint(0, 3)]
    if random.random() < weight[new_pos] / weight[pos]:
        pos = new_pos
    histo[pos] += 1

norm = sum(weight)

print('comparison:  weight, histogram')
for k in range(len(histo)):
    print('site: ', k,' weight: ', weight[k], ' histo: ', norm * histo[k] / float(n_iter))
```

```python
# pebble_multirun_histogram
xvec = {1:3, 2:2, 3:1, 4:3, 5:2, 6:1, 7:3, 8:2, 9:1}
yvec = {1:1, 2:1, 3:1, 4:2, 5:2, 6:2, 7:3, 8:3, 9:3}

neighbor = {1 : [2, 4, 1, 1], 2 : [3, 5, 1, 2], 3 : [3, 6, 2, 3],
            4 : [5, 7, 4, 1], 5 : [6, 8, 4, 2], 6 : [6, 9, 5, 3],
            7 : [8, 7, 7, 4], 8 : [9, 8, 7, 5], 9 : [9, 9, 8, 6]}

def plot_histogram(list_vec):
    x = [xvec[k] for k in list_vec]
    y = [yvec[k] for k in list_vec]

    plt.xticks([])
    plt.yticks([])
    H, xedges, yedges = np.histogram2d(x, y, bins=(3, 3),
               range=[[1,3],[1,3]], normed=True)
    print(H) # start
    print()
    H /= np.sum(H)
    print(H) # end
    extent = [yedges[0], yedges[-1], xedges[-1], xedges[0]]
    plt.imshow(H, extent=extent, interpolation='nearest', vmin=0.0, vmax=1.0)
    plt.set_cmap('hot')
    plt.colorbar()
    plt.title('t = '+str(run),fontsize=22)
    plt.show()

def run_mcmc(list_vec, run):
    for n_runs in range(100000):
        pos = 9
        for iter in range(run):
            pos = neighbor[pos][ random.randint(0, 3)]
        list_vec.append(pos)
    return list_vec

list_vec = []
run = 0
list_vec = run_mcmc(list_vec, run)
plot_histogram(list_vec)
```

```python
# pebble_multirun_all_histogram
N_runs = 10
for run in range(N_runs): # for every max move, compute its probability matrix
    list_vec = []
    list_vec = run_mcmc(list_vec, run)
    plot_histogram(list_vec)
```

### Transfer matrix

```python
neighbor =  [[1, 3, 0, 0], [2, 4, 0, 1], [2, 5, 1, 2],
             [4, 6, 3, 0], [5, 7, 3, 1], [5, 8, 4, 2],
             [7, 6, 6, 3], [8, 7, 6, 4], [8, 8, 7, 5]] 

def get_transfer(neighbor):
    """
    transfer matrix: defining the probability of transfering to all points from a certain point
    """
    transfer = np.zeros((9, 9)) 
    for k in range(9):
        for neigh in range(4): 
            transfer[neighbor[k][neigh], k] += 0.25
    return transfer

transfer = get_transfer(neighbor)

# probability vector (at time t)
position = np.zeros(9) 
position[8] = 1.0 # start point
for t in range(100):
    print(t,'  ',["%0.5f" % i for i in position])
    position = np.dot(transfer, position)
```

### The equilibrium probability vector is an eigenvector of the transfer matrix, with eigenvalue 1.

```python
# pebble_transfer_eigen
eigenvalues, eigenvectors = np.linalg.eig(transfer)
print(["%0.2f" % i for i in eigenvalues])

# you may print the eigenvectors by uncommenting the following lines...
#for iter in range(9):
#    print(eigenvalues[iter])
#    for i in range(9):
#       print(eigenvectors[i][iter])
```

```python
equilibrium_value = 1/9

position = np.zeros(9)
position[8] = 1.0
for t in range(100):
    print(t,'  ',["%0.4f" % abs(i - equilibrium_value) for i in position])
    position = np.dot(transfer, position)
```
